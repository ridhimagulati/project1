**Handling the missing values**

```python
# Replace -200 with NaN
air_quality_data = air_quality_data.replace(-200, np.NaN)
```

```python
#checking the no of missing values in the dataframe
air_quality_data.isnull().sum()
```

```python
# Calculate the mean of only the numeric columns
air_quality_data.select_dtypes(include='number').mean()
```

```python
#all the missing values will be replaced with the mean of that column
air_quality_data = air_quality_data.fillna(air_quality_data.select_dtypes(include='number').mean())
```

**Forecasting with FB prophet algorithm**

```python
# converting date from DD/MM/YYYY TO YYYY/MM/DD
date_info=pd.to_datetime(air_quality_data['Date'],format='%d/%m/%Y')
print(date_info)
```

```python
time_info=pd.to_datetime(air_quality_data['Time'],format='%H.%M.%S')
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#To remove 1900-01-01 from time column because it was converted to datetime type
time_info = time_info.dt.strftime('%H:%M:%S')
print(time_info)
```

```python
from datetime import date
#combining 2 series to a dataframe
date_time=pd.concat([date_info,time_info],axis=1)
print(date_time)
```

```python
#combining date and time
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
print(date_time)
```

```python
date_time.info()
```

We have to convert ds from object datatype to date time format

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
data.head()
```

```python
from prophet import Prophet
```

```python
#Training the prophet model
model=Prophet()
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=500,freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1=model.plot(forecast)
```
